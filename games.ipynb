В user_score есть значение tbd - должно быть определено, т.е. это значение неизвестно, поэтому заменим его на NaN.

У нас есть рейтинг:
 - E - Для всех,
 - M - Для взрослых, 
 - T - Подросткам, 
 - E10+ - Для всех от 10 лет и старше, 
 - K-A - Для всех, 
 - AO - Только для взрослых, 
 - EC - Для детей младшего возраста, 
 - RP - Рейтинг ожидается

RP - неизвестен, поэтому заменим его на NaN. Т.к K-A - старое название рейтинга E, то заменим его на него.
 А все NaN заменим на unknown.

```python
# заменим K-A на E, RP на NaN, 
data['rating'].mask(data['rating'] == "K-A", 'E', inplace=True)
data['rating'].mask(data['rating'] == "RP", np.nan, inplace=True)
data['rating'] = data['rating'].fillna('unknown')
data['rating'].unique().tolist()
```

```python
# фун-ия группировки рейтинга
def group_rating(rating): 
    '''
    E - для всех
    M, AO - для взрослых
    T, E10+ - для детей 10+
    EC - для маленьких детей
    unknown - неизвестно
    '''
    if rating == 'E':
        return 'для всех'
    if rating == 'M' or rating == 'AO':
        return 'для взрослых'
    if rating == 'T' or rating == 'E10+':
        return 'для детей 10+'
    if rating == 'EC':
        return 'для маленьких детей'
    return 'неизвестно'
```

```python
data['group_rating'] = data['rating'].apply(group_rating)

data.head()
```

#### Рассмотрим  продажи по платформам за актуальный период

```python
# фун-ия для разбивки списка на меньшие длиной size
def partition(l, size):
    for i in range(0, len(l), size):
        yield list(islice(l, i, i + size))
```

```python
# посмотрим кол-во проданных игр за последние три года
platforms = data_last_3_year['platform'].unique().tolist()
platforms = list(partition(platforms, 6))

for i in platforms:
    ax = (
        data_last_3_year
        .pivot_table(index='year_of_release', columns='platform', values='sales', aggfunc='sum')
        [i]
        .plot(figsize=(8, 6), grid=True, title='Количество проданных игр по годам')
    )
    ax.grid(alpha=0.3)
    ax.set_xlabel('год релиза')
    ax.set_ylabel('продажи')
    plt.show()
```

```python
# функция группировки данных по параметру с суммарным кол-вом продаж в различных регионах
def sales_research(df, parameter):
   return (
       (
           df
           .groupby(by=parameter)
           .agg({'na_sales': 'sum', 'eu_sales': 'sum', 'jp_sales': 'sum', 'sales': 'sum'})
           .sort_values(by='sales', ascending=False)
           .filter(regex="^(?!sales$)")
       )
   )
```

```python
# фун-ия выделения цветом максимума и минимума в табл.
def styler_sales(df):
    return (
        df.style
        .highlight_max(color='green', subset=['na_sales', 'eu_sales', 'jp_sales'])
        .highlight_min(color='coral', subset=['na_sales', 'eu_sales', 'jp_sales'])
    )
```

```python
# фун-ия группировки топ 5 и остальные 
def group_top5(df, parametr):
    top5 = df.head().index.tolist()
    all_index = df.index.tolist()

    distribution = {i: i if i in top5 else 'other' for i in all_index}
    df[f'{parametr}_groups'] = df.index.map(distribution)
    df_group = (
        df.groupby(by=f'{parametr}_groups')
        .agg({'na_sales': 'sum', 'eu_sales': 'sum', 'jp_sales': 'sum'})
    )
    return df_group
```

```python
# фун-ия построения круговой диаграммы 
def plot_pie_plot(df, title, explode):
    explode = explode
    fig, ax = plt.subplots(1, 3, figsize=(20, 13))
    fig.tight_layout(pad=5.0)
    fig.suptitle(title, fontsize=40, fontweight='bold')
    df.plot(
        kind='pie', subplots=True, figsize=(25, 10), autopct='%1.0f%%', explode=explode, 
        legend=False, fontsize=17, ax=ax, startangle=150, ylabel='',
        textprops={'color':"maroon"}
    )
    ax[0].set_title('Северная Америка', fontsize=25, loc='left', fontweight='bold', y= 1.2)
    ax[1].set_title('Европа', fontsize=25, loc='left', fontweight='bold', y=1.2)
    ax[2].set_title('Япония', fontsize=25, loc='left', fontweight='bold', y=1.2)
    plt.show()
```

### Проверка гипотезы

 1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
 2. Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Задайте самостоятельно пороговое значение alpha.

Сформулируем нулевую и альтернативную гипотезы для проверки первой гипотезы. Они должны быть взаимоисключающими, в нулевой гипотезе равенство. Альфу возьмем 0.05, т.к. данных немного. Получаем:

$\begin{equation*}
 \begin{cases}
   H_0 :\text{Средние пользовательские рейтинги платформ Xbox One и PC равны}\\
   H_1 :\text{Средние пользовательские рейтинги платформ Xbox One и PC отличаются}
 \end{cases}
\end{equation*}$

alpha = 0.05

Для начала надо проверить распределение данных. Для нормального распределения будем использовать t-тест, для распределения отличного от нормально используем непараметрический тест Манна-Уитни.

```python
# построим распределение средней пользовательской оценки для платформ Xbox One и PC
data_hypothesis_testing = data_last_3_year.dropna()
pc = data_hypothesis_testing.query('platform=="pc"')['user_score']
xone = data_hypothesis_testing.query('platform=="xone"')['user_score']


sbn.distplot(pc);
sbn.distplot(xone);
```

```python
# строим qq-plot для нормального распределения
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
st.probplot(pc, dist="norm", plot=plt)
plt.subplot(2,2,2)
st.probplot(xone, dist="norm", plot=plt)
plt.show()
```

Распределение близко к нормальному, т.к. исследуемое распределение (синяя полоска) накладывается на нормальное (красная полоска), поэтому используем t-тест.

```python
# t-тест для первой гипотезы
results = st.ttest_ind(pc, xone, equal_var=False)
alpha = .05

print('p-значение: ', results.pvalue)
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

В результате теста p-value получилось больше альфы, поэтому гипотезу средние пользовательские рейтинги платформ Xbox One и PC не отвергаем.

Для второй гипотезы рассуждения аналогичны, получим:
$\begin{equation*}
 \begin{cases}
   H_0 :\text{Средние пользовательские рейтинги жанров Action и Sports равны}\\
   H_1 :\text{Средние пользовательские рейтинги жанров Action и Sports отличаются}
 \end{cases}
\end{equation*}$

alpha = 0.05

```python
# построим распределение средней пользовательской оценки для жанров Action и Sports
action = data_hypothesis_testing.query('genre=="action"')['user_score']
sports = data_hypothesis_testing.query('genre=="sports"')['user_score']

sbn.distplot(action);
sbn.distplot(sports);
```

```python
# строим qq-plot для нормального распределения
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
st.probplot(action, dist="norm", plot=plt)
plt.subplot(2,2,2)
st.probplot(sports, dist="norm", plot=plt)
plt.show()
```

У жанра action распределение сильно отличается от нормального, однако все равно используем t-тест

```python
results = st.ttest_ind(action, sports, equal_var=False)
alpha = .05

print('p-значение: ', results.pvalue)
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

В результате теста p-value получилось меньше альфы, поэтому гипотезу средние пользовательские рейтинги жанров Action и Sports одинаковые отвергаем.

#### Вывод

При проверке гипотезы средние пользовательские рейтинги платформ Xbox One и PC использовали t-тест. Мы получили p-value больше альфы, поэтому эту гипотезу мы не отвергаем.

При проверке гипотезы средние пользовательские рейтинги жанров Action и Sports использовали  t-тест. Мы получили p-value меньше альфы, поэтому эту гипотезу мы отвергаем.
